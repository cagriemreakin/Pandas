<p style="font-family: Arial; font-size:1em;color:#2462C0; font-style:bold">
Sözlük kullanarak dataframe' lerimizi oluşturalım.</p>


```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])   
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                     index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                     index=[8, 9, 10, 11])
```

<p style="font-family: Arial; font-size:3.75em;color:purple; font-style:bold"><br>
Birbirine Bağlama (Concatenate) </p><br> 

```python
frames = [df1, df2, df3]
result = pd.concat(frames)
display (df1,df2,df3,result)
```

<p style="font-family: Arial; font-size:1em;color:#2462C0; font-style:bold">
 Yukarıdaki işlemi yaparak aşağıdaki sonuca ulaştık.Sol tarafta frame'lerin ilk halleri ve birleşrtirdikten sonra oluşan result frame'i.</p>


<p style="font-family: Arial; font-size:1em;color:#2462C0; font-style:bold">
 İstediğimiz takdirde birleştirme işlemini framelere spesifik anahtar kelime vererekte yapabiliriz. </p>


```python
result = pd.concat(frames, keys=['x', 'y', 'z'])
```

```python
display (df1,df2,df3,result)
```

```python
result.loc['z']
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],                 
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
    
```

<p style="font-family: Arial; font-size:1em;color:green; font-style:bold">
Birleştirme Çeşitleri ve Mantığı nasıldır? 
Aşağıdaki resme baktığınızda axis ile ne demek istediğimi daha iyi anlayacaksınız.</p></br>

* 1. Outer join: Bütün sutunları boş eleman dahi olsa birleştir.
* 2. İnner join: En az 2 tane DF (dataFrame)'in kesişimini alarak yeni bi DF oluşturur.
* 3. Spesifik bir index üzerinde birleştirme.

<p style="font-family: Arial; font-size:2em;color:#2462C0; font-style:bold"><br>
Outer Join Örneği</p><br> 

```python
frame = [df1,df4]
result = pd.concat(frame, axis=1)
```

```python
display (df1,df2,result)
```

<p style="font-family: Arial; font-size:2em;color:#2462C0; font-style:bold"><br>
Inner Join Örneği</p><br> 

```python
result = pd.concat(frame, axis=1, join= 'inner')
display (df1,df4,result)
```

* Gördüğünüz gibi inner join sadece 2 df'in kesişimi olan 2. ve 3. satırdaki değerleri aldı.

```python
result = pd.concat(frame, axis=0, join='inner')
display (df1,df4,result)
```

* Bu seferde sütunlara göre uyguladık

<p style="font-family: Arial; font-size:2em;color:#2462C0; font-style:bold">
Spesifik indekse göre join işlemine bakalım.</p>

```python
result = pd.concat(frame , axis=1, join_axes=[df1.index])
display (df1,df4,result)
```

* Yukarıda gördüğünüz gibi sonucu df1'ın indekslerine göre oluşturduk. df4 indeksleri 2,3 ,6 ve 7 idi. Yukarıdaki sonuca bakarsanız kesişimde sadece 2. ve 3. satırlarda df4 ün değeri olduğunu göreceksiniz.

<p style="font-family: Arial; font-size:2em;color:#2462C0; font-style:bold"><br>
Append() Fonksiyonu Kullanarak Birbirine Bağlama (Concatenate) </p><br> 

Append fonksiyonu concat fonksiyonuna göre işlevselliği ve verimliliği daha azdır. Axis= 0, yani indeksler üzerinde işlem yapar.Sırayla örneklere bakalım

```python
display (df1,df2)
result=df1.append(df2)
result
```

```python
display(df1,df4)
result = df1.append(df4)
result
```

<p style="font-family: Arial; font-size:1.5em;color:red; font-style:bold">
^*^Not: Bu işlemi yaparken df1 değerlerinde herhangi bir değişiklik olmuyor.
df1' ın kopyasına df4 değerleri eklenip result frame'inde kaydediliyor. </p>

Tabii ki de 2 frame birleştirmek zorunda değiliz. şimdi 3 tane frame'i birleştirelim. 

```python
display(df1,df2,df4)
result = df1.append([df2,df4])
result
```

<p style="font-family: Arial; font-size:2em;color:#2462C0; font-style:bold"><br>
Indeksleri Yok Sayarak Concat ve Append  </p><br> 

Farz edelimki anlamlı indeklerimiz yok aynen df4'te olduğu gibi. İstediğimiz takdirde bu indeksleri yok sayabiliriz.

```python
display(df1,df4)
result = pd.concat([df1,df4], ignore_index=True)
result
```

```python
result = df1.append(df4, ignore_index=True)
result
```
